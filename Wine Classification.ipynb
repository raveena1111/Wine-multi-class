```python
import numpy as np
from sklearn.datasets import load_wine
```

```python
df_wine = load_wine()
```

```python
x = df_wine.data
y = df_wine.target
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
dt_reg = DecisionTreeClassifier(random_state = 42)
rf_reg = RandomForestClassifier(random_state = 42)
svc_reg = SVC(random_state = 42)
```

# Decision Tree Classifier

```python
dt_reg.fit(x_train, y_train)
dt_pred = dt_reg.predict(x_test)
dt_accuracy = accuracy_score(y_test, dt_pred)
```

# Random Forest Classifier

```python
rf_reg.fit(x_train, y_train)
rf_pred = rf_reg.predict(x_test)
rf_accuracy = accuracy_score(y_test, rf_pred)
```

# SVC

```python
svc_reg.fit(x_train, y_train)
svc_pred = svc_reg.predict(x_test)
svc_accuracy = accuracy_score(y_test, svc_pred)
```

### Display

```python
print(f"Accuracy of Decision Tree Classifier: {dt_accuracy*100: .2f}%")
print(f"Accuracy of Random Forest Classifier: {rf_accuracy*100: .2f}%")
print(f"Accuracy of Support Vector Classifier: {svc_accuracy*100: .2f}%")
```

--------------
# Best Fit Algorithms
--------------

```python
best_accuracy = max(dt_accuracy, rf_accuracy, svc_accuracy)

if best_accuracy == dt_accuracy:
    best_model = "Decision Tree Classifier"
elif best_accuracy == rf_accuracy:
    best_model = "Random Forest Accuracy"
else:
    best_model = "Support Vector CLassifier"
```

```python
print(f"Best model: {best_model} with accuracy of {best_accuracy*100: .2f}%")
```
